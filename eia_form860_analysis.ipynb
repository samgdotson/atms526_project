```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('../data_misc/eia_form860/3_1_Generator_Y2020.xlsx', skiprows=1,
                   usecols=['Utility ID',
                            'Plant Code',
                            'Plant Name',
                            'Technology',
                            'State',
                            'Nameplate Capacity (MW)'])
df.head(5)
```

```python
nuclear_plants = df[df.Technology=='Nuclear']
```

```python
codes = np.array(nuclear_plants['Plant Code'].values)
names = np.array(nuclear_plants['Plant Name'].values)
states = np.array(nuclear_plants['State'].values)
```

```python
mydict = dict(zip(codes,names))
mydict2 = dict(zip(codes,states))

code_df = pd.DataFrame({'Plant Code':list(mydict.keys()),
                        'Plant Name':list(mydict.values()),
                        'State':list(mydict2.values())})
code_df.set_index('Plant Code', inplace=True)
```

```python
grouped = nuclear_plants.groupby('Plant Code').sum()
```

```python
combined = pd.concat([grouped, code_df], axis=1)
```

```python
combined.to_csv('eia_nuclear_plant_codes.csv')
```

```python
env_cooling = ['OC', 'RC', 'ON']
is_once_thru = lambda x: x in env_cooling
```

```python
cool = pd.read_excel('../data_misc/eia_form860/6_2_EnviroEquip_Y2020.xlsx', skiprows=1, sheet_name='Cooling',
                     usecols=['Plant Code',
                              'Plant Name', 
                              'State',
                              'Water Source Code',
                              'Water Type Code',
                              'Cooling Type 1',
                              'Cooling Type 2',
                              'Cooling Type 3',
                              'Cooling Type 4',
                              'Cooling Water Source',
                              'Cooling Water Discharge',
                              'Intake Rate at 100% (Gallons per Minute)'
                             ],
                     na_filter=True,
                     na_values=['',' '])
cool.head()
```

```python
np.unique(cool.iloc[:,-1].values)
```

```python
cool['discharge_to_env'] = cool["Cooling Type 1"].apply(is_once_thru)
```

```python
cool
```

```python
cool = cool.groupby('Plant Code').sum().reset_index()
cool
```

```python
total = combined.merge(cool, on='Plant Code')
```

```python
total.head()
```

```python
ot_plants = total[total['discharge_to_env']>0]
ot_plants.head()
```

```python
ot_plants.drop(columns=['Cooling Type 4', 'discharge_to_env'],inplace=True)
```

```python
ot_plants.head()
```

```python
ot_plants.rename(columns={'Intake Rate at 100% (Gallons per Minute)':'max_withdrawal_rate_kgM'},inplace=True)
```

```python
ot_plants.to_csv('nuclear_plants_OT.csv',index=False)
```
